```python
# process diacritics bahnaric
bahnaric = [unicodedata.normalize('NFC', x) for x in bahnaric]

# Print the column values
print(bahnaric)
```

```python
# remove spaces at the beginning and end of the words
bahnaric = [x.strip() for x in bahnaric]

# if in a word there is ", " or " ,", replace it with ","
bahnaric = [x.replace(", ", ",") for x in bahnaric]
bahnaric = [x.replace(" ,", ",") for x in bahnaric]
```

```python
def partition(bahnaric_list):
    # if there is a space, split the word at the space
    new_bahnaric = []
    for word in bahnaric_list:
        if " " not in word:
            # split comma
            if "," in word:
                word = word.split(",")
                for words in word:
                    new_bahnaric.append(words)
            else:
                new_bahnaric.append(word)

        else:
            word = word.split(" ")
            # print(word)
            for words in word:
                # print(words)
                if "," in words:
                    words = words.split(",")
                    for wordss in words:
                        new_bahnaric.append(wordss)
                else:
                    new_bahnaric.append(words)
    
    # pass by value new_bahnaric list to return_bahnaric
    return_bahnaric.extend(new_bahnaric)
```

```python
# process bahnaric list by function partition
partition(bahnaric)
print(return_bahnaric)
```

```python
# A full pattern of Bahnaric language: Cp1VCp2Cm(1,2)WVmCf
# Cp1: consonant before pre_vowel: Containing p,t,c,k,q,b,d,g,đ,s,h,m,n,nh,ng,w,l,r,j
# V: pre_vowel: i,a,ơ
# Cp2: consonant after pre_vowel: Containing m n nh ng l
# Cm1: consonant before main_vowel, containing b, c, d, đ, g, h, k, l, m, n, p, q, r, s, t, v, x, ch, gh, gi, kh, ng, ngh, nh, th, tr, qu
# Cm2: containing l, r, y
# W: containing u or o, but then the previous character is q then W must be associated with u
# Vm is main_vowel: i, ĭ, iê, iê̆, ê, ê̆, e, ĕ, ư, ư̆, ươ, ươ̆, ơ, ơ̆, a, ă, u, ŭ, uô, uô̆, ô, ô̆, o, ŏ
# Cf: Can also be Cm1 or W
```

```python
Cp1 = ["p", "t", "c", "k", "q", "b", "d", "g", "đ", "s", "h", "m", "n", "w", "l", "r", "j", "nh", "ng", "ngh", "ph", "gh", "th", "ch", "kh", "sh", "qh", "sr", "sl"]
V = pre_vowels
Cp2 = ["m", "n", "nh", "ng", "l"]
Cm1 = ["b", "c", "d", "đ", "g", "h", "k", "m", "n", "p", "q", "s", "t", "v", "x", "ch", "gh", "gi", "kh", "ng", "ngh", "nh", "th", "tr", "qu"]
Cm2 = ["l", "r", "y"]
Vm = main_vowels
Cf = Cm1
```

```python
# normalize by unicodedata
Cp1 = [unicodedata.normalize('NFC', i) for i in Cp1]
V = [unicodedata.normalize('NFC', i) for i in V]
Cp2 = [unicodedata.normalize('NFC', i) for i in Cp2]
Cm1 = [unicodedata.normalize('NFC', i) for i in Cm1]
Cm2 = [unicodedata.normalize('NFC', i) for i in Cm2]
Vm = [unicodedata.normalize('NFC', i) for i in Vm]
Cf = [unicodedata.normalize('NFC', i) for i in Cf]
```

```python
def cluster_word(list):
    for i in list:
        print(i)
        Cp1_word = ""
        Cp2_word = ""
        V_word = ""
        Cm1_word = ""
        Cm2_word = ""
        Vm_word = ""
        Cf_word = ""

        cluster_A = ""
        cluster_B = ""

        store_pre = -1
        store_main = -1
        j = 0
        index = 0
        break_check = 0

        while index in range(len(i)):
            if i[index] in V:
                if index + 1 < len(i):
                    if i[index + 1] == special_char:
                        index += 2
                        continue
                
                    else:
                        store_pre = index
                        break
                
                else:
                    store_pre = index
                    break
            
            else:
                index += 1
                continue

        if store_pre != -1:
            gap = store_pre
            while (gap < len(i)):
                if (i[gap] in V):
                    
                    gap += 1
                    continue
                
                else:
                    break

            index = gap
        
            while index in range(gap, len(i)):
                # print(i[index])
                if i[index] in Vm:
                    store_main = index
                    break
                else:
                    index += 1
                    continue
        
        else:
            index = 0
            while index in range(store_pre + 1, len(i)):
                # print(i[index])
                if i[index] in Vm:
                    store_main = index
                    break
                else:
                    index += 1
                    continue

        if store_pre == -1:
            # no pre_vowel? --> Cp1 and Cp2 in empty
            if store_main == 0:
                Cm1_word = ""
                Cm2_word = ""
            else:
                Cm1_word = i[0:store_main]
                Cm2_word = ""
            index = 0
            for index in range(len(i)):
                if i[index] in Vm:
                    while index < len(i):
                        print(index)
                        if i[index] == special_char or i[index] in Vm:
                            Vm_word += i[index]
                            print(Vm_word)
                            index += 1
                            continue
                        else:
                            break
                    break_check = 1
                if break_check == 1:
                    break

            # Cf word is the rest, meaning i - Vm_word - Cm1_word
            # Cf must be taken from tail up to the place where there is no Vm_word
            final = len(i) - 1
            while final > index - 1:
                if i[final] not in Vm_word:
                    Cf_word = Cf_word + i[final]
                    final -= 1
                    continue
                else:
                    break

        elif store_main == -1:
            # no main_vowel? --> Cm1 and Cm2 in empty
            Cp1_word = i[0:store_pre]
            for index in i:
                if index in pre_vowels:
                    V_word = index
                    break
            final = len(i) - 1
            while True:
                if i[final] not in V_word:
                    Cp2_word = Cp2_word + i[final]
                    final -= 1
                    continue
                else:
                    break

        else:
            print(store_pre)
            print(store_main)
            j = 0
            while j in range(len(i)):
                #print(i[j])
                if i[j] == special_char:
                    j += 1
                    continue
                # pre_vowels can have up to 2 characters. We should make sure that
                # the program detects 2 characters as 1 pre_vowel (if that group of
                # 2 consecutive characters are in pre_vowels). Otherwise, we consider
                # single character pre_vowel as normal.
                if j < len(i) - 1:
                    if j == store_main:
                        break
                    if i[j:j+2] in pre_vowels:
                        V_word += i[j:j+2]
                        j += 2
                        continue

                    else:
                        print(i[j])
                        if i[j] in pre_vowels:
                            if i[j:j+2] in main_vowels:
                                j += 2
                                continue
                            
                            elif j-1 >= 0:
                                if i[j-1:j+1] in main_vowels:
                                    if i[j-1] == special_char:
                                        if i[j-2:j+1] in main_vowels:
                                            j += 1
                                            continue

                                        else:
                                            V_word += i[j]
                                            j += 1
                                            continue
                                    else:
                                        j += 1
                                        continue

                                else:
                                    V_word += i[j]
                                    j += 1
                                    continue
                            
                            else:
                                V_word += i[j]
                                j += 1
                                continue
                            

                        else:
                            #print(i[j])
                            #print(store_pre)
                            #print(store_main)
                            if j < store_pre:
                                if i[j] in Cp1:
                                    Cp1_word += i[j]
                                j += 1
                                continue

                            elif j > store_pre and j < store_main:
                                
                                if i[j] in Cp2 and i[j] in Cm1:
                                    Cm1_word += i[j]

                                elif i[j] in Cp2 and i[j] in Cm2:
                                    Cm1_word += i[j]

                                elif i[j] in Cp2:
                                    Cp2_word += i[j]

                                # elif i[j] in Cm1:
                                #     Cm1_word += i[j]

                                # elif i[j] in Cm2:
                                #     Cm2_word += i[j]

                                j += 1
                                continue

                            elif j == store_main:
                                j += 1
                                continue
                            
                            elif j > store_main:
                                if i[j] in Cm1:
                                    Cm1_word += i[j]

                                elif i[j] in Cm2:
                                    Cm2_word += i[j]

                                j += 1
                                continue

                elif j == store_main:
                    break

                else:
                    j += 1
                    continue

        cluster_A = Cp1_word + V_word + Cp2_word
        # cluster B is the difference between cluster A and i
        j = 0
        for j in range(len(cluster_A), len(i)):
            cluster_B += i[j]


        print([cluster_A, cluster_B])
        cluster_list.append(cluster_A)
        cluster_list.append(cluster_B)
```
